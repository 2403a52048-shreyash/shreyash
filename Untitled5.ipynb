```python
def is_leap_year(year):
  """Checks if a given year is a leap year."""
  if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    return True
  else:
    return False

# Example usage (not part of the requested zero-shot prompt)
# print(is_leap_year(2024))
# print(is_leap_year(1900))
# print(is_leap_year(2000))
```

```python
Task Description#3
● Few-shot: Provide 2–3 examples to generate a function that formats full names as
“Last, First”.
Expected Output#3
● Well-structured function respecting the examples
```

```python
def format_name(full_name):
  """Formats a full name as 'Last, First'."""
  # Examples:
  # "John Doe" -> "Doe, John"
  # "Alice Wonderland" -> "Wonderland, Alice"
  # "Peter Pan" -> "Pan, Peter"
  parts = full_name.split()
  if len(parts) > 1:
    return f"{parts[-1]}, {' '.join(parts[:-1])}"
  else:
    return full_name # Return original if no space found
```

## Generate zero-shot code

### Subtask:
Generate the Python code for a function that counts vowels in a string using a zero-shot prompt.


**Reasoning**:
Write a Python function that counts the number of vowels (a, e, i, o, u, case-insensitive) in a given string using a zero-shot prompt.



```python
def count_vowels_zero_shot(input_string):
  """Counts the number of vowels in a string (zero-shot)."""
  vowels = "aeiouAEIOU"
  vowel_count = 0
  for char in input_string:
    if char in vowels:
      vowel_count += 1
  return vowel_count
```

## Generate few-shot code

### Subtask:
Generate the Python code for a function that counts vowels in a string using a few-shot prompt (providing examples).


**Reasoning**:
The subtask requires generating a Python function to count vowels using a few-shot approach with examples in the docstring, so I will write the function definition with the specified examples and logic.



```python
def count_vowels_few_shot(input_string):
  """
  Counts the number of vowels in a string (few-shot).

  Examples:
  "hello" -> 2
  "Programming" -> 3
  "AEIOU" -> 5
  """
  vowels = "aeiouAEIOU"
  vowel_count = 0
  for char in input_string:
    if char in vowels:
      vowel_count += 1
  return vowel_count

print("Function `count_vowels_few_shot` defined.")
```

**Reasoning**:
The function has been defined. The next step is to execute the function with some test cases to ensure it works correctly based on the provided examples.

```python
ask Description#5
● Use few-shot prompting to generate a function that reads a .txt file and returns the
number of lines.
Expected Output#5
● Working file-processing function with AI-guided logic
```

```python
# Create a dummy text file for demonstration
with open("sample.txt", "w") as f:
  f.write("This is the first line.\n")
  f.write("This is the second line.\n")
  f.write("And the third.\n")
  f.write("The fourth line.\n")
  f.write("The fifth and final line.\n")

print("Created 'sample.txt' with 5 lines.")
```

```python
def count_lines_in_file(filename):
  """
  Reads a text file and returns the number of lines.

  Examples:
  If 'sample.txt' contains:
  Line 1
  Line 2
  Line 3
  count_lines_in_file('sample.txt') should return 3.

  If 'empty.txt' is an empty file, count_lines_in_file('empty.txt') should return 0.
  """
  try:
    with open(filename, 'r') as f:
      lines = f.readlines()
      return len(lines)
  except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
    return -1 # Or raise an exception, depending on desired error handling
```

```python
# Test the function with the created file
line_count = count_lines_in_file("sample.txt")
if line_count != -1:
  print(f"The file 'sample.txt' has {line_count} lines.")

# Test with a non-existent file
count_lines_in_file("non_existent_file.txt")
```
